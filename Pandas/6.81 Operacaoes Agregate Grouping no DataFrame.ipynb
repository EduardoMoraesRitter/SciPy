```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('primary-results.csv')
df.head()
```

```python
df.groupby('candidate')
```

```python
df.groupby('candidate').aggregate({'votes':[min, np.mean, max]})
```

```python
df[df['votes'] == 590502]
#ou
#df.loc([df['votes'] == 590502])
```

```python
df.groupby('candidate').aggregate({'fraction_votes':[min, np.mean, max]})
```

```python
df[df['fraction_votes'] == 1]
```

```python
df[(df['fraction_votes'] == 1) & (df['candidate'] == 'Hillary Clinton')]
```

```python
"""
ou
df.gruopby('state').filter(x['fraction_votes'] > 0.8)
"""
def fraction_votes_filter(x):
    return x['votes'].sum() > 300000
df.groupby('state').filter(fraction_votes_filter)
```

```python
df[(df['state_abbreviation'] == 'AL')]['votes'].sum()
#filta e soma
```

```python
def fraction_votes_filter(x):
    return x['votes'].sum() > 1000000
df.groupby('state').filter(fraction_votes_filter)
```

```python
df.groupby(['state_abbreviation', 'candidate'])['votes'].sum()
#agrupado por dois, e soma
```

```python
#fazer o soma dos votos, transendo somente os acima da media
```

```python
df.groupby('state').filter(lambda x : x['votes'].sum() > 300000)
```

```python
#df[(df['candidate'].str.contains("Trump|Hillary "))]
df[(df['candidate'] == "Hillary Clinton") | (df['candidate'] == "Donald Trump")]
```
