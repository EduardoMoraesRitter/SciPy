```python
import numpy as np
import matplotlib.pyplot as pl
%matplotlib inline
```

```python
salarios_du = np.array([100,200,400,300])
salarios_jo = np.array([50,300,500,450])
```

```python
#label
pl.plot(salarios_du, c='Black', ls='--', marker='s', ms=8, label='Salario Du')
pl.plot(salarios_jo, c='Red', ls='--', marker='s', ms=8, label='Salario Jo')
pl.show()
```

```python
#legenda
pl.plot(salarios_du, c='Black', ls='--', marker='s', ms=8, label='Salario Du')
pl.plot(salarios_jo, c='Red', ls='--', marker='s', ms=8, label='Salario Jo')
pl.legend()
pl.show()
```

```python
pl.plot(salarios_du, c='Black', ls='--', marker='s', ms=8, label='Salario Du')
pl.plot(salarios_jo, c='Red', ls='-', marker='^', ms=8, label='Salario Jo')
pl.legend(loc='lower right')
pl.show()
```

```python
salario_pedro = [200,150,300,700]
pl.plot(salarios_du, c='Black', ls='--', marker='s', ms=10, label='Salario Du')
pl.plot(salarios_jo, c='Red', ls='-', marker='^', ms=10, label='Salario Jo')
pl.plot(salario_pedro, c='Green', ls='-', marker='o', ms=8, label='Salario Pedro')
pl.legend(loc='lower right')
pl.show()
```
