```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
%matplotlib inline             #nao esquecer
```

```python
salario = np.array([100,200,300,500, 400, 150])
salario
```

```python
plt.plot(salario)
```

```python
plt.plot(salario, c='black')
```

```python
plt.plot(salario, ls='--')
```

```python
plt.plot(salario, ls='--', marker='s')
```

```python
plt.plot(salario, c='black', ls='--', marker='s', ms=18)
plt.show()
```
