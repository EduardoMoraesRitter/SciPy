```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.plot(data[:50,0], c='Red', ls=':', marker='s', ms=8)
```

```python
#50 a 99
plt.plot(data[50:100,0], c='Blue', ls=':', marker='s', ms=8)
```

```python
#o comprimento da sepala de versicolor > Setosa
plt.plot(data[0:50,0], c='Red', ls=':', marker='s', ms=8, label='Iris-Setosa')
plt.plot(data[50:100,0], c='Blue', ls=':', marker='s', ms=8, label='Iris=Versicolor')
plt.legend()
plt.show()
```

```python
#comprimento sepala
plt.plot(data[0:50,0], c='Red', ls=':', marker='s', ms=8, label='Iris-Setosa')
plt.plot(data[50:100,0], c='Blue', ls=':', marker='s', ms=8, label='Iris=Versicolor')
plt.plot(data[100:150,0], c='Green', ls=':', marker='s', ms=8, label='Iris=Outra')
plt.legend()
plt.show()
```

```python
#largura sepala
plt.plot(data[0:50,1], c='Red', ls=':', marker='s', ms=8, label='Iris-Setosa')
plt.plot(data[50:100,1], c='Blue', ls=':', marker='s', ms=8, label='Iris=Versicolor')
plt.plot(data[100:150,1], c='Green', ls=':', marker='s', ms=8, label='Iris=Outra')
plt.legend()
plt.show()
```
